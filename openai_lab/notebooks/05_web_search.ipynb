# Exercise 5: Web Search Tool in the Responses API

The Responses API supports built-in tools. Here we use the `web_search` tool  
to let the model fetch live information from the web, then inspect the output  
structure including search call items and citation annotations.

```python
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv()
client = OpenAI()
```

## Call the API with the web_search tool enabled

```python
response = client.responses.create(
    model="gpt-4.1-mini",
    tools=[{"type": "web_search"}],
    input="What were OpenAI's most significant product announcements in the last 30 days?",
)
```

## Display the response text

```python
print("=== Response ===\n")
print(response.output_text)
```

## Inspect the output items and citations

The output contains both `web_search_call` items and `message` items with annotations.

```python
print("=== Output items ===")
for i, item in enumerate(response.output):
    print(f"\n[{i}] type={item.type}")
    if item.type == "web_search_call":
        print(f"    id={item.id}")
        print(f"    status={item.status}")
    elif item.type == "message":
        for j, content in enumerate(item.content):
            if content.type == "output_text":
                # Show annotations (citations)
                if content.annotations:
                    print(f"    Citations found: {len(content.annotations)}")
                    for k, ann in enumerate(content.annotations):
                        print(f"    [{k}] {ann.type}: {ann.title}")
                        print(f"         URL: {ann.url}")
```

## Token usage

```python
print("=== Usage ===")
print(f"Input:  {response.usage.input_tokens} tokens")
print(f"Output: {response.usage.output_tokens} tokens")
```

## Interview one-liner

> **Built-in tools like `web_search`** let the model autonomously fetch live data mid-response; the output includes structured citation annotations so you can verify sources and build grounded, up-to-date answers without managing a separate search pipeline.
