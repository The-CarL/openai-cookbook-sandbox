# Exercise 1: Basic Responses API Call

The simplest possible interaction with OpenAI's Responses API.  
We send a single prompt, get a response, and then inspect the full response structure
including metadata, token usage, and the output item hierarchy.

```python
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv()
client = OpenAI()
```

## Send a single prompt to the Responses API

```python
response = client.responses.create(
    model="gpt-4.1-mini",
    input="What are the three most important things an AI Solutions Engineer should understand about LLM APIs? Be concise.",
)
```

## Quick access to the response text

```python
print("=== Quick access ===")
print(response.output_text)
```

## Inspect response metadata

```python
print("=== Response metadata ===")
print(f"ID:             {response.id}")
print(f"Model:          {response.model}")
print(f"Status:         {response.status}")
print(f"Usage (input):  {response.usage.input_tokens} tokens")
print(f"Usage (output): {response.usage.output_tokens} tokens")
print(f"Usage (total):  {response.usage.total_tokens} tokens")
```

## Explore the output structure

The response output is a list of output items. Each message item contains content blocks.

```python
print("=== Output structure ===")
for i, item in enumerate(response.output):
    print(f"Output[{i}]: type={item.type}")
    if item.type == "message":
        for j, content in enumerate(item.content):
            print(f"  Content[{j}]: type={content.type}, text={content.text[:80]}...")
```

## Interview one-liner

> **`response.output_text`** is the fast path to the answer, but the full `response` object gives you token counts, model info, and the structured output hierarchy -- all essential for production observability.
